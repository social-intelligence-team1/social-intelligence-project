```python
def classify_region(place):
    place = str(place)

    for keyword, region in region_map.items():
        if keyword in place:
            return region

    return '미분류'
```

```python
illegal['지역'] = illegal['단속장소'].apply(classify_region)
illegal
```

```python
#미분류 행 몇개
(illegal['지역'] == '미분류').sum()
```

```python
unclassified = illegal[illegal['지역'] == '미분류']
unclassified['단속장소'].value_counts()
```

```python
#미분류 된 데이터 삭제
illegal = illegal[illegal['지역'] != '미분류']
```

```python
region_count_df = illegal['지역'].value_counts().reset_index()

region_count_df.columns = ['지역', '단속건수']

region_count_df
```

```python
illegal
```

```python
import plotly.express as px

region_df = illegal['지역'].value_counts().reset_index()
region_df.columns = ['지역', '단속건수']

fig = px.bar(
    region_df,
    x='지역',
    y='단속건수',
    text='단속건수',
    title='전체 지역별 불법주정차 단속 건수'
)

fig.update_layout(
    xaxis_title='지역',
    yaxis_title='단속 건수',
    xaxis_tickangle=-45,
    width=1200,
    height=600
)

fig.show()
```

```python
area = pd.read_excel('../data/raw/세종시_동별_면적.xlsx')
area
```

```python
merged_df = pd.merge(
    region_count_df,
    area,
    on='지역'
)

merged_df['면적 당 단속건수'] = (
    merged_df['단속건수'] / merged_df['면적']
).round(0).astype(int)

merged_df

```

```python
merged_df = merged_df.sort_values(
    by='면적 당 단속건수',
    ascending=False
)

merged_df
```

```python
fig = px.bar(
    merged_df,
    x='지역',
    y='면적 당 단속건수',
    text='면적 당 단속건수',
    title='전체 지역별 불법주정차 단속 건수'
)

fig.update_layout(
    xaxis_title='지역',
    yaxis_title='면적 당 단속 건수',
    xaxis_tickangle=-45,
    width=1200,
    height=600
)

fig.show()
```

```python
merged_df.to_csv('../data/processed/illegal_summary.csv', index=False, encoding='utf-8-sig')
```

```python
# 아름동,조치원읍,도담동,소담동,보람동,나성동,한솔동만 뽑음
target_regions = ['아름동', '조치원읍', '도담동', '소담동', '보람동', '나성동', '한솔동']

illegal_selected = illegal[illegal['지역'].isin(target_regions)]
illegal_selected['지역'].value_counts()
illegal_selected
```

```python
top_places_count = (
    illegal_selected
    .groupby('지역')['단속장소']
    .apply(lambda x: x.value_counts().head(5))
)
top_places_count_df = top_places_count.reset_index()
top_places_count_df.columns = ['지역', '단속장소', '단속건수']
top_places_count_df.to_csv('../data/processed/illegal_top_places.csv', index=False, encoding='utf-8-sig')
```
