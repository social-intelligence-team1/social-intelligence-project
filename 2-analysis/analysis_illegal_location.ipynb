```python
import folium

nasung_map = folium.Map(location=[36.483, 127.26], zoom_start=16)

df_nasung = df[(df['지역'] == '나성동') & df['위도'].notna() & df['경도'].notna()]

for _, row in df_nasung.iterrows():
    radius_scale = row['단속건수'] * 0.05
    popup_text = f"{row['단속장소']}<br><b>단속건수:</b> {row['단속건수']}건"
    
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=radius_scale,            # 단속건수별 크기 동적 조절
        popup=folium.Popup(popup_text, max_width=300),
        color='red',                    # 원의 테두리 색상
        fill=True,                      # 원 내부 채우기 여부
        fill_color='red',               # 원 내부 채우기 색상
        fill_opacity=0.4                # 내부 투명도 (겹치는 구역 시인성 확보)
    ).add_to(nasung_map)

parking_filtered = parking_locations[parking_locations['위도'].notna() & parking_locations['경도'].notna()]

for _, row in parking_filtered.iterrows():
    parking_radius = row['주차구획수'] * 0.05
    
    if parking_radius < 4:
        parking_radius = 4
        
    parking_name = row.get('주차장명', '주차시설')
    popup_text = f"<b>[주차장]</b> {parking_name}<br><b>주차구획수:</b> {row['주차구획수']}면"
    
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=parking_radius,
        popup=folium.Popup(popup_text, max_width=300),
        color='blue',                   
        fill=True,
        fill_color='blue',
        fill_opacity=0.4
    ).add_to(nasung_map)

# 지도 출력
nasung_map.save('../results/nasung_maps.html')
```
